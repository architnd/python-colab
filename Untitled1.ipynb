```python
# A list of fruits
fruits = ["apple", "banana", "cherry"]

# Adding a new item
fruits.append("dragonfruit")

# Logic to find fruits with more than 5 letters
for fruit in fruits:
    if len(fruit) > 5:
        print(f"{fruit.title()} is a long word!")
    else:
        print(f"{fruit.title()} is a short word.")
```

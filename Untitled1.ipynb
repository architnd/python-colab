```python
# A list of fruits
fruits = ["apple", "banana", "cherry"]

# Adding a new item
fruits.append("dragonfruit")

# Logic to find fruits with more than 5 letters
for fruit in fruits:
    if len(fruit) > 5:
        print(f"{fruit.title()} is a long word!")
    else:
        print(f"{fruit.title()} is a short word.")
```

```python
fruits = {1:"apple", 2:"banana", 3:"cherry", 4:"starwberry", 5:"grape", 6:"dragonfruit"}
for i in fruits:
  print(i, fruits[i])
for x,y in fruits.items(): # Corrected to iterate over key-value pairs
  print(x,y)
print(list(fruits.keys()))
print(list(fruits.values()))
print(list(fruits.items()))
keys = list(fruits.keys())
values = list(fruits.values())
print(zip(keys, values))
fruits2 = dict(zip(keys, values))
print(fruits2)
fruits3 = {x:y for x,y in zip(keys, values)}
print(fruits3)
fruits_ = {7:"apple", 8:"banana", 9:"cherry", 10:"starwberry", 11:"grape", 12:"dragonfruit"}
fruits4 = {**fruits, **fruits_}
print(fruits4)
fruits5 = {x:y for x,y in fruits.items() if x > 3}
print(fruits5)
fruits6 = {x:y for x,y in fruits.items() if 'rry' in y}
print(fruits6)
fruits7 = {x:y for x,y in fruits.items() if x % 2 == 0}
print(fruits7)
fruits8 = {i:i**2 for i in range(1,11)}
print(fruits8)
print(fruits8.items())
fruits9 = dict(filter(lambda x : x[0] % 2 == 0, fruits8.items()))
print(fruits9)
fruits10 = dict(map(lambda x : (x[0], x[1]**2), fruits8.items()))
print(fruits10)
```

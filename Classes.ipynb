```python
class student:
    studentCount = 0
    __studentCount = 0
    def __init__(self, name, age, major):
        self.name = name
        self.age = age
        self.major = major
        student.studentCount += 1
        student.__studentCount += 1
    def __del__(self):
        print("Object is deleted")
        student.studentCount -= 1
        student.__studentCount -= 1
    def greet(self):
        return f"Hello, my name is {self.name}, I am {self.age} years old and I am majoring in {self.major}."

class child(student):
    def __init__(self, name, age, major, parent):
        super().__init__(name, age, major)
        self.parent = parent
    def __del__(self):
        return super().__del__()
    def greet(self):
        return f"Good morning, my name is {self.name}, I am {self.age} years old and I am majoring in {self.major}."

# Create an instance of the student class
student1 = student("Alice", 20, "Computer Science")
child1 = child("Bob", 15, "English", "John")
# Call the greet method
print(student1.greet())
print(child1.greet())
print(student1.studentCount)
#print(student._student__studentCount)
# Delete the instance
del student1
del child1
#print(student.studentCount)
#print(student._student__studentCount)
```

```python
class User:
  def __init__(self, name, age, email):
    self.name = name
    self.age = age
    self.email = email

a = User("archit", 20, "architnd")
print(a.name)
print(a.age)
print(a.email)
b = User("archit", "abc", "architnd") #this is allowed
print(b.name)
print(b.age)
print(b.email)
```

```python
from pydantic import BaseModel, ValidationError, StrictInt
class User(BaseModel):
  name : str
  #age : int
  age : StrictInt
  email : str
try:
  a = User(name = "archit", age = 20, email = "architnd")
  print(a.name)
  #b = User(name = "archit", age = "abc", email = "architnd") #this is not allowed
  #print(b.name)
  b = User(name = "archit", age = "20", email = "architnd")
  print(b.name)
except ValidationError as e:
  print(e)
```

```python
from pydantic import BaseModel, ValidationError, conint, constr, field_validator
class User(BaseModel):
  name : constr(min_length=3, max_length=10)
  age : conint(ge=10, le=60)
  email : str
  @field_validator("email")
  @classmethod
  def check_email(cls, v):
    if "@" not in v:
      raise ValueError("Invalid email")
    return v
try:
  a = User(name = "archit", age = 20, email = "architnd")
  print(a.name)
  #b = User(name = "archit", age = 5, email = "architnd@")
  #print(b.name)
except ValidationError as e:
  print(e)
```

```python
from pydantic import BaseModel, ValidationError, model_validator

class User(BaseModel):
  age : int
  @model_validator(mode="before")
  def check_age_present(cls, data):
    if "age" not in data:
      raise ValueError("Age is required")
    return data
print(User(age=20))
try:
  a = User()
except ValidationError as e:
  print(e)
```

```python
from pydantic import BaseModel, ValidationError, model_validator

class User(BaseModel):
  name : str
  age : int

  @model_validator(mode="after")
  def check_age_range(self) -> 'User':
    if self.age < 18:
      raise ValueError("Age must be 18 or greater")
    return self

print(User(name="Alice", age=20))

try:
  a = User(name="Bob", age=15)
except ValidationError as e:
  print(e)
```

```python
from pydantic import BaseModel
class User(BaseModel):
  name : str
  age : int
user = User(name="Alice", age=20)
json_str = user.model_dump_json()
print(json_str)
```

```python
from pydantic import validate_call, ValidationError
@validate_call
def add(age: int, name: str):
  return f"Hello, my name is {name} and I am {age} years old."
print(add(20, "Alice"))
try:
  print(add("Alice", 20))
except ValidationError as e:
  print(e)
```
